#Metodo de la Biseccion

```python
#Metodo de Biseccion
def Biseccion(Lower,Upper):
  #Copiamos los limites
  a=Lower
  b=Upper

  #Establecemos los basicos
  Epsilon=0.001
  Contador=0
  Registro=[]


  #Iniciamos el metodo
  while(np.abs(a-b)>Epsilon):
    alpha=float((a+b)/2)

    #Obtenemos las derivadas
    DFa=float(EvalDeriv(a))
    DFalpha=float(EvalDeriv(alpha))

    #Alteramos los valores de su siguiente recursion
    if(DFa*DFalpha<0):
      b=alpha
    else:
      a=alpha

    #Evaluamos a el candidato
    Fa=float(EvalFunction(a))
    Registro.append([Contador,a,Fa])
    Contador+=1

    #imprimimos los datos
    print("* Iteracion",(Contador),"  - X:",float(a),"  - Y:",float(Fa))


  return Registro
```

#Metodo de Newton-Raphson

```python
#Metodo de newton-Raphson
def NewtonRaphson(Inic):

  #Declaramos punto de inicio
  x=Inic

  #establecemos nociones previas
  Previo=1000
  Epsilon=0.001

  Contador=0
  Registro=[]

  #Iniciamos metodo
  while(np.abs(x-Previo)>Epsilon):
    #obtenemos sus dos derivadas
    D1x=float(EvalDeriv(x))
    D2x=float(EvalSecDeriv(x))

    #Establecemos el nuevo valor de x
    Previo=x
    x=Previo-(D1x/D2x)

    #Evaluamos
    Fx=float(EvalFunction(x))


    #Agregamos a registro
    Contador+=1
    Registro.append([Contador,x,Fx])
    print("* Iteracion",Contador,"  - X:",float(x),"  - Y:",float(Fx))

  return Registro
```

#Metodo de la Secante

```python
#Metodo de la Secante
def Secante(Lower,Upper):
  #Pasamos los limites
  a=Lower
  b=Upper

  #Establecemos nociones previas
  D1alpha=100
  Epsilon=0.001

  Contador=0
  Registro=[]


  #Empezamos metodo
  while(np.abs(D1alpha)>Epsilon):
    #Obtenemos las evaluaciones de sus derivadas
    D1a=EvalDeriv(a)
    D1b=EvalDeriv(b)

    #Obtenemos a alpha
    alpha=float(b-(D1b/((D1b-D1a)/(b-a))))

    #Vemos su derivada
    D1alpha=float(EvalDeriv(alpha))

    #Dependiendo del resultado, cambiamos a a o b
    if(D1alpha>0):
      b=alpha
    else:
      a=alpha

    #Evaluamos al punto
    Falpha=float(EvalFunction(alpha))

    #Agregamos al registro
    Contador+=1
    Registro.append([Contador,alpha,Falpha])

    print("* Iteracion",Contador,"  - X:",float(alpha),"  - Y:",float(Falpha))

  return Registro
```

#Metodo de Busqueda de la Seccion Dorada

```python
def Dorada(Lower,Upper):
  #Establecemos los limites
  a=Lower
  b=Upper

  #Establecemos nociones previas
  Phi=1.618033988
  Tau=2-Phi
  Epsilon=1e-6

  Contador=0
  Registro=[]

  Falpha1=100
  Falpha2=0

  #Iniciamos el metodo
  while(np.abs(Falpha1-Falpha2)>Epsilon):
    #obtenemos a alpha 1 y alpha 2
    alpha1=a*(1-Tau)+b*Tau
    alpha2=a*Tau+b*(1-Tau)

    #Los evaluamos
    Falpha1=EvalFunction(alpha1)
    Falpha2=EvalFunction(alpha2)

    #Si alphax resulta dar mas grande, se cambia el valor de x
    if(Falpha1>Falpha2):
      a=alpha1
    else:
      b=alpha2

    #Agregamos al registro
    Contador+=1
    Registro.append([Contador,alpha2,Falpha2])
    print("* Iteracion",Contador,"  - X:",float(alpha2),"  - Y:",float(Falpha2))

  return Registro
```

#Metodo de Fibonacci

```python
def Fibonacci(Lower,Upper,n):
  #Estavlecemos limites
  a=Lower
  b=Upper
  k=2


  #Establecemos nociones previas
  Registro=[]
  Contador=0
  Epsilon=1e-6

  Fxa=100
  Fxb=0
  L=b-a

  #Obtenemos un array con la serie de los numeros de Fibonacci
  Serie=SerieFib(n)

  #Iniciamos el Metodo
  while(np.abs(Fxa-Fxb)>Epsilon):

    #Obtenemos Lk con los penultimos valores de la serie
    Lk=(Serie[n-k+1]/Serie[n+1])*L

    #Obtenemos a xa y xb
    xa=a+Lk
    xb=b-Lk

    #Los evaluamos
    Fxa=EvalFunction(xa)
    Fxb=EvalFunction(xb)

    #Vemos quien es mas grande de Fxy y y se cambia
    if(Fxa>Fxb):
      a=xa
    else:
      b=xb

    #Aumentamos recursion
    k+=1

    #Agregamos al registro
    Contador+=1
    Registro.append([Contador,xa,Fxa])

    print("* Iteracion",Contador,"  - X:",float(xa),"  - Y:",float(Fxa))


  return Registro

def SerieFib(n):
  fs=[1,1]
  for i in range(n):
    F=fs[-1]+fs[-2]
    fs.append(F)
  return fs
```

#Concluciones

Los Metodos anteriormente descritos nos ayudan a encontrar maximos y minimos en una funcin, sin embargo, dichos mximos y mnimos estarn atados a diversos elementos, por ejemplo, de donde se parta para encontrarlos.

As mismo, se debe notar que no tenemos control sobre a cual maximo y minimo converge el metodo, si no que se decanta por los algoritmos usados, siendo que a primeras instancias no podemos declarar los elementos como globales o locales.

Sin embargo un punto adestacar es que podemos manipular las condiciones iniciales para que busque la convergencia del problema, siendo as que podemos mover las acotaciones iniciales o el punto inicial (en el caso de Newton-Raphson) para favorecer su convergencia hacia cierto punto crtico.

Adems cabe aclarar sobre las diferencias de los mtodos, iniciando con el Mtodo de Biseccion, el cual converge a un ritmo promedio y sus saltos no son tan grandes entre generaciones, seguido del de Newton-Raphson, que es muy ideal, puesto que a pesar que sus saltos son grandes entre generaciones, converge de manera veloz a un punto critico. Ademas de que al solo tener un punto inicial, puede explorar todo el espacio con mayor libertad, ventaja que comparte con el metodo de la Secante. Por otro lado, los metodos que no se atienen a los gradientes como el de Busqueda de la Seccion Dorada (BSD) y Fibonacci, comparten la cualidad de que se quedan en el espacio de busqueda, lo que ayuda a que no se alejen de la seccin que queramos rebuscar, util si solo queremos analizar dicho espacio de busqueda en vez de todo el espacio.

En conclucin, cada mtodo tiene sus cualidades, cada uno util en casos especificos, por lo que debemos tener a consciencia sus cualidades para saber cual sera conveniente utilizar en diversas y especficas situaciones.
